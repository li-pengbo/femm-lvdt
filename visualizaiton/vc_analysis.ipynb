```python
import sys
sys.dont_write_bytecode = True
sys.path.append('../')
import json
from modules import dataFrame, dataHandler
import numpy as np
import scienceplots
import scipy.optimize as opt
import matplotlib.pyplot as plt
import random
%config InlineBackend.figure_format = 'svg'

datafolder_path = "../data"
VC_data_files = dataFrame.VC_data_files
Magnet_data_files = dataFrame.Magnet_data_files
```

```python
datapaths= {'magnet core':'/alucoil_vc/VC_magcore_d8_l10.h5',
            'alu' : '/core_material_vc/VC_alcld_id08_od14_l12_magcore_d8_l10.h5',
            'supermalloy' : '/core_material_vc/VC_supermalloy_magnetcore_d8_l10.h5',
            'pureiron': '/core_material_vc/VC_pureiron_magnetcore_d8_l10.h5'
            }

dataset= {}
for key in datapaths:
    data = dataHandler.data_preprocessing(dataHandler.load_data(datafolder_path + datapaths[key]))
    dataset[key] = data
for key in dataset:
    data = dataset[key]
    plt.plot(data['CC_pos'], data['M_force'], marker = 'o', markersize = 8, label='Magnet Force + ' + key)
    plt.xlabel('CC Position (mm)')
    plt.ylabel('Force (N)')
    plt.legend(loc = 'center')
    plt.show()
for key in dataset:
    data = dataset[key]
    plt.plot(data['CC_pos'], abs(data['OC_low_force'] + data['OC_upp_force']), marker = 'o', markersize = 8, label='Outercoil Force + ' + key)
    plt.xlabel('CC Position (mm)')
    plt.ylabel('Force (N)')
    plt.legend(loc = 'center')
    plt.show()
```

```python
jsonfilepath = '../json/vc_data_magnet_length.json'
data = dataHandler.get_data_from_json(datafolder_path, jsonfilepath)
VC_alucld_d08d07_data = data['VC_alucld_id07_od08']
VC_alucld_d08d06_data = data['VC_alucld_id06_od08']
VC_alucld_d10d09_data = data['VC_alucld_id09_od10']
VC_alucld_d10d08_data = data['VC_alucld_id08_od10']
```
